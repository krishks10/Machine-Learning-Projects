```python
import zipfile

with zipfile.ZipFile('StoreSaleArchive.zip', 'r') as zip_ref:
    zip_ref.extractall('SaleData')

print("Files extracted successfully.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('SaleData/train.csv')
print(data.head())
```

```python
data.head()
```

```python
data.info()
```

```python
data["Postal Code"].fillna(0, inplace=True)
data["Postal Code"] = data["Postal Code"].astype(int)
```

```python
data.info()
```

```python
type_of_costumer = data['Segment'].unique()
print(type_of_costumer)
```

```python
no_of_costumers = data['Segment'].value_counts().reset_index()
no_of_costumers = no_of_costumers.rename(columns = {'Segment': 'Type of Customers'})
print(no_of_costumers)
```

```python
plt.pie(no_of_costumers['count'], labels=no_of_costumers['Type of Customers'], autopct='%1.1f%%')
plt.title('Distribution of Customer Segments')
```

```python
data.head(3)
```

```python
sales_per_segment = data.groupby('Segment')['Sales'].sum().reset_index()
print(sales_per_segment)
```

```python
plt.bar(sales_per_segment['Segment'], sales_per_segment['Sales'])
plt.xlabel('Customer Segment')
plt.ylabel('Total Sales')
plt.title('Sales by Customer Segment')
```

```python
customer_order_frequency = data.groupby(['Customer ID', 'Customer Name', 'Segment'])['Order ID'].count().reset_index()
customer_order_frequency.rename(columns = {'Order ID': 'Total Orders'}, inplace = True)
```

```python
repeat_customers = customer_order_frequency[customer_order_frequency['Total Orders'] >= 1]
print(repeat_customers)
```
